```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
from pathlib import Path
from tensorflow import keras
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
```

```python
MODEL_PATH = "/content/drive/MyDrive/RNN/final_efficientnetb0.keras"
IMG_SIZE = (224, 224)
THRESHOLD = 0.5
CLASS_NAMES = ["fire", "no_fire"]

model = load_model(MODEL_PATH)
print("Modèle chargé avec succès depuis :", MODEL_PATH)

def predict_one_image(img_path, show_image=True):
    img = keras.utils.load_img(img_path, target_size=IMG_SIZE)
    arr = keras.utils.img_to_array(img)
    arr = np.expand_dims(arr, axis=0).astype("float32")

    prob_no_fire = float(model.predict(arr, verbose=0)[0][0])
    prob_fire = 1.0 - prob_no_fire

    if prob_no_fire >= THRESHOLD:
        pred_idx = 1
        pred_name = CLASS_NAMES[pred_idx]
        confidence = prob_no_fire
    else:
        pred_idx = 0
        pred_name = CLASS_NAMES[pred_idx]
        confidence = prob_fire

    print("Image :", img_path)
    print(f"Probabilité no_fire : {prob_no_fire:.10f}")
    print(f"Probabilité fire    : {prob_fire:.10f}")
    print("Classe prédite      :", pred_name)
    print(f"Confiance           : {confidence:.10f}")

    if show_image:
        plt.figure(figsize=(4, 4))
        plt.imshow(img)
        plt.title(f"{pred_name} | conf={confidence:.4f}")
        plt.axis("off")
        plt.show()

    return {
        "prob_no_fire": prob_no_fire,
        "prob_fire": prob_fire,
        "prediction": pred_name,
        "confidence": confidence
    }
```

```python
img_path = "/content/drive/MyDrive/RNN/images/fire.739.png"
result = predict_one_image(img_path)
print(result)
```
